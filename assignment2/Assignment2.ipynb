```python
from datetime import datetime
from dateutil import parser
import numpy
```

```python
len(df)
```

```python
def parse_datetime(x):
    try:
        return parser.parse(x)
    except:
        return numpy.nan
df['datetime'] = df['Date'].apply(parse_datetime)
```

```python
df
```

```python
import seaborn as sns
ax = sns.scatterplot(x="High", y="Low",hue='Difference',data=df)
```

```python
df.set_index('datetime').resample('M').aggregate('max').plot()
```
